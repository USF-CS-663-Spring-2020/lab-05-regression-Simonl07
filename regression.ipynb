```python
import pandas as pd
df = pd.read_csv('toluca.csv')
df
```

```python
df.plot(kind='scatter', x='lotSize', y='workHours')
```

```python
import seaborn as sns
plot = sns.lmplot(y='workHours', x='lotSize', data=df)
```

```python
r1 = sns.residplot('lotSize', 'workHours', data=df)
r1.set(xlabel='Lot Size', ylabel='Residuals')
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df.drop('workHours', axis=1), df.iloc[:,1])
lin_reg
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
import pandas as pd
df = pd.read_csv('../Credit.csv')
df
```

```python
df.plot(kind='scatter', x='Limit', y='Income')
df.plot(kind='scatter', x='Limit', y='Rating')
df.plot(kind='scatter', x='Limit', y='Cards')
df.plot(kind='scatter', x='Limit', y='Age')
df.plot(kind='scatter', x='Limit', y='Education')
```

```python
import seaborn as sns
plot = sns.lmplot(y='Limit', x='Income', data=df)
plot = sns.lmplot(y='Limit', x='Rating', data=df)
plot = sns.lmplot(y='Limit', x='Cards', data=df)
plot = sns.lmplot(y='Limit', x='Age', data=df)
plot = sns.lmplot(y='Limit', x='Education', data=df)
```

```python
r1 = sns.residplot('Limit', 'Income', data=df)
r1.set(xlabel='Limit', ylabel='Residuals')
```

```python
from sklearn.linear_model import LinearRegression

predictors = ['Income', 'Rating', 'Cards', 'Age', 'Education']

for predictor in predictors:
    lr = LinearRegression().fit(df.filter(items=['Limit']), df.loc[:, predictor])
    print(f'{predictor}: coef: {lr.coef_}, intercept: {lr.intercept_}')
```
